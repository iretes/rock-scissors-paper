# Voters model

\begin{cases}
R + S \xrightarrow{P_r} 2\cdot R\\
S + P \xrightarrow{P_s} 2\cdot S\\
P + R \xrightarrow{P_p} 2\cdot P\\
P \xrightarrow{P_t} R
\end{cases}

```python
import stochpy
```

```python
!cat ../long_range_models/voters_model.psc
```

```python
smod = stochpy.SSA(IsQuiet=False)
smod.Model(model_file="voters_model.psc",dir="../long_range_models/")
smod.ShowSpecies()
```

```python
smod.DoStochSim(IsTrackPropensities=True, end=1000000)
smod.PlotSpeciesTimeSeries()
```

```python
smod.ChangeParameter("Pt", 0)
smod.DoStochSim(IsTrackPropensities=True, end=1000000)
smod.PlotSpeciesTimeSeries()
```
